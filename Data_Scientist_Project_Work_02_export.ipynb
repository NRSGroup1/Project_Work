```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load data using Pandas
df_crm_model = pd.read_csv('crm_model.csv', sep=',')
df_crm_model
```

```python
#Load data using Pandas
df_finance_model = pd.read_csv('finance_model.csv', sep=',')
df_finance_model
```

```python
#Load data using Pandas
df_sales_model = pd.read_csv('sales_model.csv', sep=',')
df_sales_model
```

```python
#Load data using Pandas
df_twitter_model = pd.read_csv('twitter_model.csv', sep=',')
df_twitter_model
```

```python
#Show the complete pandas dataframe
pd.set_option("display.max_colwidth", -1)

#Load data definitions using Pandas
df_dataset_definition = pd.read_csv('dataset_definition.csv', sep=',')
df_dataset_definition
```

Explore crm_model

```python
df=df_crm_model
```

```python
#Dataset infos
df_info=df_dataset_definition[(df_dataset_definition.dataset == 'CRM')]
df_info
```

```python
#Shape
df.shape
```

```python
#Check for uniquenes of Id (Column 0, duplicated result: False or True); Wanted: Only False
df[df.columns[0]].duplicated().max()
```

```python
#Datatypes
df.info(verbose=True)
```

```python
#Number Variables (float64 and int64): basic stats
df.describe()
```

```python
#Number Variables (float64 and int64): Countplot
object_bol = df.dtypes == 'float64'
for catplot in df.dtypes[object_bol].index:
    sns.countplot(y=catplot,data=df)
    plt.show()   
    
object_bol = df.dtypes == 'int64'
for catplot in df.dtypes[object_bol].index:
    sns.countplot(y=catplot,data=df)
    plt.show()
```

```python
#Character Variables (object): Countplot
object_bol = df.dtypes == 'object'
for catplot in df.dtypes[object_bol].index:
    sns.countplot(y=catplot,data=df)
    plt.show()
```
